```python
#setup
import csv
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import seaborn as sns
```

```python
cars_file = r"C:\Users\lnovo_crisissupport\Desktop\SMU Data Analytics BC\HW\Project 1\Project-1\USA_cars_datasets.csv"
df=pd.read_csv(cars_file)

# Show Data
cars_file
```

```python
clean_csv = df.dropna()
clean_csv
```

```python
brand_mean  = df.groupby('brand')['price'].mean()
brand_mean
```

```python
plt.figure(figsize=(12,6))
sorted_nb = df.groupby(['brand'])['price'].mean().sort_values()
plt.title('Cost Distribution of Cars Across Different Brands (Ascending Order)')
sns.boxplot(x=df['brand'], y=df['price'], order=list(sorted_nb.index))
plt.xticks(rotation=70)
```

```python
brand_mean_sorted = brand_mean.sort_values()

plt.figure(figsize=(10, 6))
plt.bar(brand_mean_sorted.index, brand_mean_sorted.values, color='skyblue')
plt.xlabel('Car Brand')
plt.ylabel('Mean Cost')
plt.title('Mean Cost of Cars Across Different Brands (Ascending Order)')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
